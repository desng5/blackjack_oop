```python
import random #will shuffle deck before dealing


#control flow / Global variables 
#tups
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
#dict
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

#will contrl while loops
playing = True 
```

```python
class Card:
    
    def __init__(self, suit, rank): 
        self.suit = suit #type of card
        self.rank = rank #rank of card
        
    def __str__(self):
        return self.rank + ' of ' + self.suit #concatenate string to rep type of card/rank

        
```

```python
class Deck: 
    
    def __init__(self):#initializing deck w/suit & rank from card class
        self.deck = [] #empty list to append to
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
            
    def __str__(self):
        in_deck = ''  # empty string
        for card in self.deck: #loops through entire deck
            in_deck += '\n '+card.__str__() 
        return 'The deck has:' + in_deck #returns/print string of cards in deck.
    
    
    def shuffle(self):
        random.shuffle(self.deck) #inplace shuffle
        
    def deal(self):
        return self.deck.pop()
        
```

```python
class Hand:
    
    #initializing start of hand and what cards are in hand
    def __init__(self): 
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self,card):
        self.cards.append(card) # call from method above - deck.deal add to self.cards[]
        self.value += values[card.rank] #value dict and rank add it to self.value count
        if card.rank == 'Ace':
            self.aces += 1 #adds to self.aces count

    
    def adjust_for_aces(self):
        
        #if value is >21 with an ace in hand, change ace value to 1 from 11
        while self.value > 21 and self.aces > 0:
            self.value -= 10 #subt from value if True
            self.aces -= 1 #subt from aces if True
        
```

```python
class Chips:
    
    def __init__(self):
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
def display_chips(chips):
    print("Available chips:", chips.total)
```

```python
def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input('What would you like to bet? '))
        except ValueError:
            print('Sorry, your bet needs to be an integer!')
        else:
            if chips.bet > chips.total:
                print("Sorry, bet exceeds available", chips.total)
            else:
                break
            
```

```python
#func will be called when player hits or dealer has less than 17
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_aces()
    
def hit_or_stay(deck,hand):
    global playing #controls the while loops of this block
    
    while True:
        x = input("Do you want to Hit or Stay? Enter 'h' or 's' ")
        
        if x[0].lower() == 'h':
            hit(deck,hand) #if player selects 'h', hit() func will come run.
            
        elif x[0].lower() == 's':
            print("Player stays. Dealer plays.")
            playing = False
            
        else: 
            print("Sorry, please try again.")
            continue
        break
```

```python
#will display cards during play
def show_cards(player,dealer):
    print("\nDealer's Hand:")
    print(" <hidden>") #will keep dealers first card hidden
    print('',dealer.cards[1])  #shows dealers 2nd card
    print("\nPlayer's Hand:", *player.cards, sep='\n ') #sep= paramater that will separate and add new line
                                                        # * will print every card...
def show_all_cards(player,dealer):
    print("\nDealer's Hand:", *dealer.cards, sep='\n ')
    print("Dealer's Hand =",dealer.value)
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print("Player's Hand =",player.value)
```

```python
def player_blackjack(player,dealer,chips):
    print("Player BLACKJACK!")
    player_chips.total += player_chips.bet

def player_busts(player,dealer,chips):
    print("Player BUSTS!")
    player_chips.total -= player_chips.bet

def player_wins(player,dealer,chips):
    print("Player WINS!")
    player_chips.total += player_chips.bet

def dealer_busts(player,dealer,chips):
    print("Dealer busts!")
    player_chips.total += player_chips.bet

def dealer_wins(player,dealer,chips):
    print("Dealer wins!")
    player_chips.total -= player_chips.bet

def push(player,dealer,chips):
    print("TIE! It's a push.")
```

```python
name = input("What's your name? ")

while True:
    # Print an opening statement
    print(f'Welcome to BlackJack, {name.title()}! Place your bet!')

    
    # make and shuffle deck, deal out 2 cards to each
    deck = Deck()
    deck.shuffle()
    
    player_hand = Hand()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
            
    
    player_chips = Chips()  # players chips set to default value of 100    
    
    # Ask for players bet
    take_bet(player_chips)
    
    # Show cards (but keep one dealer card hidden)
    show_cards(player_hand,dealer_hand)
    
    while playing:  # call hit or stay func again
        
        hit_or_stay(deck,player_hand) 
        
        show_cards(player_hand,dealer_hand)  
        
        # break loop if player bust
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
            break        


    # if player hasnt hit 21 or dealer plays until 17
    if player_hand.value <= 21:
        
        while dealer_hand.value < player_hand.value:
            hit(deck,dealer_hand)    
    
        # Show cards
        show_all_cards(player_hand,dealer_hand)
        
        # other ways to win
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chips)

        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)

        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)

        else:
            push(player_hand,dealer_hand)        
    
    # players chips total 
    print("\nPlayer's wallet has",player_chips.total)
    
    # Ask to play again
    new_game = input("Would you like to play another hand? Enter 'y' or 'n' ")
    
    if new_game[0].lower()=='y':
        playing=True
        continue
    else:
        print(f"Thanks for playing, {name.title()}!")
        break
```
